```python
class TicTacToe:
    def __init__(self) -> None:
        self.board = [[' ']*3 for _ in range(3)]
        self.player = 'O'

    def printBoard(self):
        for   i in self.board:
            print('|'.join(i))
            print("-----")

    def isBoardFull(self):
        return all( self.board[i][j]!=' ' for i in range(3) for j in range(3))

    def getEmptyCells( self):
        return [[i, j] for i in range(3) for j in range(3) if self.board[i][j] == ' ']

    def isWinner( self, player):
        for i in range(3) :
            if all( self.board[i][j]==player for j in range(3)) or all(self.board[j][i] == player for j in range(3)) :
                return True
        if( all( self.board[i][i]==player for i in range(3)) or all( self.board[i][2-i]==player for i in range(3))) : 
            return True
        return False
    
    def isGameOver(self):
        if self.isWinner('X') :
            print("AI wins. Better Luch Next Time!!!")
            return True
        elif self.isWinner('O'):
            print("Congratulations!!! You Won")
            return True
        elif self.isBoardFull():
            print("Tied!! ")
            return True
        else:
            return False
    def minimax( self, d, maximizer):
        if self.isWinner( 'X'):
            return 1
        elif self.isWinner('O'):
            return -1
        elif self.isBoardFull():
            return 0
        else:
            if maximizer:
                max_val=float('-inf')
                for mv in self.getEmptyCells():
                    self.board[mv[0]][mv[1]]='X'
                    max_val = max(max_val, self.minimax(d+1,False))
                    self.board[mv[0]][mv[1]]=' '
                return max_val
            else:
                min_val=float('inf')
                for mv in self.getEmptyCells():
                    self.board[mv[0]][mv[1]] = 'O'
                    min_val=min( min_val, self.minimax(d+1, True))
                    self.board[mv[0]][mv[1]] = ' '
                return min_val
            
    def dfs(self):
        bv=float('-inf')
        bm=None
        for mv in self.getEmptyCells():
            self.board[mv[0]][mv[1]] = 'X'
            val = self.minimax(0,False)
            self.board[mv[0]][mv[1]] = ' '
            if( val > bv):
                bv=val
                bm=mv

        self.board[bm[0]][bm[1]] = 'X'
        print("AI has Moved ")
        self.printBoard()
        


    
    
```
